```python
# required data
years = list(range(2006, 2025))
effects = ['Transcription factors', 'Gene expression', 'Protein structure', 'Alternative splicing', 'Gene regulatory network', 'Epigenetics']
```

```python
search_details = []
for effect in effects:
    for year in years:
        query = f'"machine learning"[All Fields] AND "predicting"[All Fields] AND "{effect}"[All Fields] AND ("{year}/01/01"[PubDate] : "{year}/12/31"[PubDate])' #creating a search key
        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        driver = webdriver.Chrome(options=chrome_options)
        driver.get("https://www.ncbi.nlm.nih.gov/pmc/?term=%22machine+learning%22%5BAll+Fields%5D+AND+%22predicting%22%5BAll+Fields%5D+AND+%22Epigenetics%22%5BAll+Fields%5D+AND+(%222024%2F01%2F01%22%5BPubDate%5D+%3A+%222024%2F12%2F31%22%5BPubDate%5D)")
        search_box = driver.find_element(By.ID, "term")
        search_box.clear()
        search_box.send_keys(query)
        search_box.send_keys(Keys.RETURN)
        page_content = driver.page_source
        soup = BeautifulSoup(page_content, 'html.parser')
        page_content = driver.page_source
        driver.quit()
        filters = soup.find_all('a', {'title': 'Total Results'})
        for filter_item in filters:
          result = filter_item.text
          res = re.sub(r'\D', '', result)
          #print(result)
          search_details.append([query, effect, year, res])
```

```python
columns = ['query', 'effect', 'year', 'Num_of_articles']
data = pd.DataFrame(search_details, columns = columns)
```

```python
data = data.rename(columns={'year': 'Year', 'Num_of_articles': 'Count'})
```
